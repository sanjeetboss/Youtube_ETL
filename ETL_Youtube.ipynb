```python
import pandas as pd
from sqlalchemy import create_engine
from pandas.io.json import json_normalize
import json
```

```python
## we will first read all the csv files and load them into data frames.
us_file = "./USvideos.csv"
gb_file = "./GBvideos.csv"
ca_file = "./CAvideos.csv"
de_file = "./DEvideos.csv"
```

```python
###converting the CSV files into df so that we can start with pre processing of the data
us_df=pd.read_csv(us_file)
gb_df=pd.read_csv(gb_file)
ca_df = pd.read_csv(ca_file)
de_df = pd.read_csv(de_file)
```

```python
us_df.head()

##as we can see that this does not have country name in it , so we will add column country name in it.
```

```python
###now we will import the json files that contains the data for the categories
us_json='./US_category_id.json'
ca_json='./CA_category_id.json'
gb_json='./GB_category_id.json'
de_json='./DE_category_id.json'
```

```python
###converting the json files into pandas dataframe
us_json_df=pd.read_json(us_json)
ca_json_df=pd.read_json(ca_json)
gb_json_df=pd.read_json(gb_json)
de_json_df=pd.read_json(de_json)
type(us_json)
```

```python
us_json_df.head()
ca_json_df.head()
```

```python
##adding country name in trending file dataframes
new_us_df = us_df[['video_id', 'trending_date', 'title','channel_title', 'category_id', 'publish_time', 'views', 'likes','dislikes', 'comment_count'  ]].copy()
new_us_df.insert(1,'country','US')
new_us_df.head()

##now we will do this for all the other countries that we have as well.

new_ca_df = ca_df[['video_id', 'trending_date', 'title','channel_title', 'category_id', 'publish_time', 'views', 'likes','dislikes', 'comment_count'  ]].copy()
new_ca_df.insert(1,'country','CA')
new_us_df.head()

new_de_df = de_df[['video_id', 'trending_date', 'title','channel_title', 'category_id', 'publish_time', 'views', 'likes','dislikes', 'comment_count'  ]].copy()
new_de_df.insert(1,'country','DE')
new_de_df.head()

new_gb_df = gb_df[['video_id', 'trending_date', 'title','channel_title', 'category_id', 'publish_time', 'views', 'likes','dislikes', 'comment_count'  ]].copy()
new_gb_df.insert(1,'country','GB')
new_gb_df.head()
```

```python
#we need to clean this and extract the categorty so that we can category id in our final trending stats
us_details=[]

for item in us_json_df["items"]:
    title= item['snippet']['title']
    id=int(item['id'])
    us_details.append({'title':title, 'id':id})
    
gb_details=[]

for item in gb_json_df["items"]:
    title= item['snippet']['title']
    id=int(item['id'])
    gb_details.append({'title':title, 'id':id})
    
de_details=[]
for item in de_json_df["items"]:
    title = item["snippet"]["title"]
    id = int(item["id"])
    de_details.append({"title":title, "id":id})
    
ca_details=[]
for item in ca_json_df["items"]:
    title = item["snippet"]["title"]
    id = int(item["id"])
    ca_details.append({"title":title, "id":id})
```

```python
##this containts information abot the category 

us_details[:5]
```

```python
###Now we are updating the data into our new data frames
us_json_df_clean=pd.DataFrame(us_details)
us_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)

gb_json_df_clean=pd.DataFrame(gb_details)
gb_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)

ca_json_df_clean=pd.DataFrame(ca_details)
ca_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)

de_json_df_clean=pd.DataFrame(de_details)
de_json_df_clean.rename(columns = {"id":"category_id","title":"category_title"},inplace=True)


us_json_df_clean.head()
ca_json_df_clean.head()
```

```python
print (new_us_df['category_id'].dtypes)
```

```python
# merging the csv df with the cleaned json df
us_merged_df = pd.merge(new_us_df ,us_json_df_clean , how = "left", on = ["category_id","category_id"])
gb_merged_df = pd.merge(new_gb_df ,gb_json_df_clean , how = "left", on = ["category_id","category_id"])
ca_merged_df = pd.merge(new_ca_df ,ca_json_df_clean , how = "left", on = ["category_id","category_id"])
de_merged_df = pd.merge(new_de_df ,de_json_df_clean , how = "left", on = ["category_id","category_id"])
```

```python
us_merged_df.head()
```

```python
###now we will merge all the country df together and then convert them into csv
final_df=pd.concat([us_merged_df,ca_merged_df,gb_merged_df,de_merged_df])
final_df = final_df.reset_index(drop=True)
final_df.insert(0,'ID',range(1,1+len(final_df)))
final_df=final_df.set_index('ID')

final_df.head()
```

```python
##now we will convert it to csv , then we can load it to our mongoDB
final_df.to_csv('./final.csv')
```
